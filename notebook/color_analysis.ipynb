```python
import os
import numpy
from pathlib import Path
from PIL import Image
from random import randint
import numpy as np
import glob
import matplotlib.pyplot as plt
import statistics 
import threading 
import scipy
from scipy.stats import linregress
```

```python
data_dir = Path('/Users/work/Data/Wheat') 
croped_dir = Path('/Users/work/Desktop/croped')

file_list = glob.glob(f"{croped_dir}/*")
```

```python
colors =  np.empty((0, 3), dtype=np.uint8)

i = 0
for element in file_list:
                image_array = np.array(Image.open(element))
                image_array = np.stack( (image_array[:, :, 0], image_array[:, :, 1], image_array[:, :, 2]), axis = 2)[ :, 0 ]
                colors = np.append(colors, image_array, axis = 0)
```

```python
plt.subplot(3, 1, 1)
plt.title("histogram of Blue")
plt.hist(colors[:, 2], color="blue", bins = 255)
  
plt.subplot(3, 1, 2)
plt.title("histogram of Green")
plt.hist(colors[:, 1], color="green", bins = 255)
  
plt.subplot(3, 1, 3)
plt.title("histogram of Red")
plt.hist(colors[:, 0], color="red", bins = 255)
  
# for clear view
plt.tight_layout()
plt.show()
  
# combined histogram
plt.title("Histogram of all RGB Colors")
plt.hist(colors[:, 2], color="blue", bins = 255)
plt.hist(colors[:, 1], color="green", bins = 255)
plt.hist(colors[:, 0], color="red", bins = 255)
plt.show()
```
